#### coefficent of determination

fitting a curve means relating an indipendent variable to a predicted value of a dependent variable

mean square error is not a good guage of how good the best fit is, hence there is the coefficent of determination



- Yi = ith observed value
- Pi be the corr predicted value
- U be the mean

 If R2 = 1, the model is a perfect fit to the data. If R2 = 0, there is no relationship between the values predicted by the model and the way the data is distributed around the mean.

this has an upper bound and a lower bound giving us an accurate way to determine how good a fit is
