#### clustering


dividing by the size of the cluster
- normalizing the values
- penalty for big cluster is the same as the small cluster
- not normalizing penalises the big clusters more than the small clusters
- the big clusters have alot of difference between the values

most common constraint
- limit the number of clusters
- try and fit the differences inside the clusters



#### hiearchial clustering
- start by assigning each item to a clusters, so have N items, N clusters each containing one item
- find the closest pair of clusters and merge, one fewer cluster
- repeat till stopping criteria is attainted

single linkage
- consider the dt between two clusters to be equal to shortest dt from any member of one cluster to any member of the other cluster

complete linkage
- consider the dt between two clusters to be equal to greatest dt from any member of one cluster to any member of the other cluster

the type of linkage can determine what kind the final clustering is
