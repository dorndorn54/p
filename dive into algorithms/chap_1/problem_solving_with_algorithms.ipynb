## Problem solving with algorithms

```python
# plotting a trajectory of a ball using galilean model

def ball_trajectory(x):
    location = 10*x - 5*(x**2)
    return(location)

ball_trajectory(5)
```

```python
# plotting the ball trajectory using matplotlib

import matplotlib.pyplot as plt
xs = [x/100 for x in list(range(201))]
ys = [ball_trajectory(x) for x in xs]
plt.plot(xs,ys)
plt.title('The Trajectory of a Thrown Ball')
plt.xlabel('Horizontal Position of Ball')
plt.ylabel('Vertical Position of Ball')
plt.axhline(y = 0)
plt.show()
```

```python
# now plot the same graph but with the line of sights of a person catching the ball
xs2 = [0.1,2]
ys2 = [ball_trajectory(0.1),0]
xs3 = [0.2,2]
ys3 = [ball_trajectory(0.2),0]
xs4 = [0.3,2]
ys4 = [ball_trajectory(0.3),0]
plt.title('The Trajectory of a Thrown Ball - with Lines of Sight')
plt.xlabel('Horizontal Position of Ball')
plt.ylabel('Vertical Position of Ball')
plt.plot(xs,ys,xs2,ys2,xs3,ys3,xs4,ys4)
plt.show()
```

### analysis

the position of the where the person stands would dictate the acceleration of the ball they see

- stand closer to the landing position and the acceleration is positive
- stand further from the landing position and the acceleration is negative
- if at the right spot acceleration of the ball is const
