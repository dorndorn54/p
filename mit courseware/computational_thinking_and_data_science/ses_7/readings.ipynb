### sampling and confidence intervals
often it is not possible to sample an entire population hence a subset is often chosen

however the sampled subset must be indicative of the population, else no amount of fancy math can make it accurate.

#### probability sampling
each member of population of interest has some non zero prob of being included in the sample


##### simple random sampling
- each member of the population has an equal chance of being chosen for the sample

##### stratified sampling
- population partitioned into subgroups
- sample is then built randomly sampling from each subgroup
- this increases the probability that a sample is representative of the population as a whole 

sample mean is a subset of the population mean

estimating the width of the confidence to achieve a desired confidence level is trickier, dependent on sample size

as the population grows
- sample mean and standard dev will be closer to the population mean and population standard dev will be closer to the real value

how big the sample size is dependent on the variance of the population


####  central limit theorm
possible to use a single sample drawn from a population to estimate the variability of the means of a set of hypothetical samples drawn from the same population

- given a set of sufficently large samples drawn from same populatiuon, the means of the samples will be approx normally distributed
- normal distribution close to the mean of the population 
- the variance of sample means close to the variance of the population divided by sample size

#### standard error of the mean
a way to estimate a confidence interval using a single example

SE for a sample of size n is the standard dev of the mean of an infinite number of samples of size n drawn from the same population.




if a sample size is large enough the diff between standard dev and population standard dev is relatively small.
the sample standard dev is a good indication of the popilation standard dev.

- able to compute the mean and standard dev of sample
- estimate SE from sample standard dev
- use SE to generate confidence intervals around sample mean
